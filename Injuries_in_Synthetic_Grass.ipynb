# Injury analysis

This code was the first atempt to answer a dilemma in the brazilian first division league.

Three Stadiums in the First division League installed a synthetic grass (Nilton Santos Stadium, Allianz Arena and Ligga Arena).

The main reason to install a synthetic grass field for those teams was a financial purpose. It is possible to host a music concert or any entertainment event one day before the soccer match and with that you may add a new type of revenue for the club. Also, the maintenance expenses may reduce with this type of grass.

However, there is this big complain about physical conditions against those fields. Some players says that the ball runs way too fast but the players feet fix really easy on the ground. Luisito Suarez declined once to attend to match in Nilton Santos Stadium, he says the field is way too hard, this makes a great impact on his knee.


Let's check if we find any evidence about this physical conditions.

1 - In Sofascore API we have a category in the substitution type that says 'injury'.


2 - We are going to use this label to count how many injury substitution we had during the 2023 Season in Brazil for each Arena.


3 - After that, we are going to calculate the rate of those substitutions.

First loop to scrap the data for of the rounds in League. All match IDs in the same list. We are doing 38 requests, for good practices I used a 2 second sleep time between those requests.

```python
ids = []
for i in range(1, 39):
  url_rodada = f"https://api.sofascore.com/api/v1/unique-tournament/325/season/48982/events/round/{i}"
  response = requests.request("GET", url_rodada , headers = headers, data = payload)
  df = pd.DataFrame(response.json()['events'])
  ids.extend(df['id'])
  time.sleep(2)
```

The second loop, we are going to run the IDs list and now get two requests per IDs. Defined 2 seconds per requests, so in total 4 seconds per match IDs. We have 38 rounds and 10 matches per round. We have 380 IDs in the list. It takes sometime.

```python

df_info = pd.DataFrame()
df_subs = pd.DataFrame()

for id in ids:
  print(id)
  time.sleep(2)
  url_partida = f"https://api.sofascore.com/api/v1/event/{id}"
  response = requests.request("GET", url_partida , headers = headers)
  partida = response.json()['event']
  df_temp = pd.json_normalize(partida)
  df_temp = df_temp[['id', 'season.id', 'roundInfo.round','venue.city.name', 'venue.stadium.name',  'venue.id']]
  df_info = pd.concat([df_info, df_temp])

  time.sleep(2)

  url_subs = f"https://api.sofascore.com/api/v1/event/{id}/incidents"
  response = requests.request("GET", url_subs , headers = headers)
  partida = response.json()['incidents']
  df_temp = pd.json_normalize(partida)
  df_temp['id_event'] = df_temp['id']
  df_temp['id'] = df_temp['id'].apply(lambda x: id)
  df_temp = df_temp[['id', 'id_event', 'incidentType','incidentClass' ]]
  df_temp = df_temp[(df_temp['incidentType'] == 'substitution') & (df_temp['incidentClass'] == 'injury')]

  df_subs = pd.concat([df_subs, df_temp])

```

This is a simple compilation time, we merge the data we get from the API.

```python

df = df_info.merge(df_subs, on='id', how='left')
df['substitutions'] = df['id_event'].fillna(0).astype(int)
df['games'] = df['id'].fillna(0).astype(int)
df['substitutions_per_game'] = df['substitutions'] / df['games']
df
```

```python

dfper_venue = df_info.merge(df_subs, on='id', how='left')
dfper_venue[dfper_venue['venue.stadium.name']=='Estadio Nilton Santos'] = 'Estádio Nilton Santos'
dfper_venue
```

Now we calculate the rate and sort by it.

```python
unique_id_per_stadium = dfper_venue.groupby('venue.stadium.name')['id'].nunique().reset_index()
unique_event_per_stadium = dfper_venue.groupby('venue.stadium.name')['id_event'].nunique().reset_index()
unique = unique_id_per_stadium.merge(unique_event_per_stadium, on='venue.stadium.name', how='left')
unique['per game'] = unique['id_event']/unique['id']
unique = unique[unique['id']>2]
unique.sort_values(by='per game', ascending=False).reset_index()
unique = unique[['venue.stadium.name','id','id_event', 'per game']]
unique = unique.rename(columns={'venue.stadium.name':'stadium','id': 'games', 'id_event': 'injuries' })
unique.sort_values(by='per game', ascending=False).reset_index()
```
